```python
def custom_title_case(name):
    words = name.split()
    title_words = []

    for word in words:
        if word.lower() == "and":
            title_words.append(word.lower())
        else:
            title_words.append(word.capitalize())

    return " ".join(title_words)


df_renamed['State/UT'] = df_renamed['State/UT'].apply(custom_title_case)


df_renamed
```

```python
def extract_acronyms_and_names(acronyms_and_names):
    acronyms = []
    names = []
    
    for entry in acronyms_and_names:
        parts = entry.split('(')
        acronyms.append(parts[0].strip())
        if len(parts) > 1:
            names.append(parts[1].strip().rstrip(')'))
        else:
            names.append('')
    
    return acronyms, names

# List of acronyms and names
acronyms_and_names = [
    'Number of Primary Health Centers(PHC, ',
    'Community Health Centers(CHC, ',
    'Sub-District/Divisional Hospitals(SDH',
    'District Hospitals(DH '
]

full_names,acronyms  = extract_acronyms_and_names(acronyms_and_names)

print(acronyms)
print(full_names)
```

```python
def remove_outliers_iqr(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return data[(data >= lower_bound) & (data <= upper_bound)]
```

```python
df["Bed_Deficit"] = remove_outliers_iqr(df["Bed_Deficit"])
```

```python
plt.figure(figsize=(12, 6))
plt.bar(df["State/UT"], df["Bed_Deficit"], color='orange')
plt.xlabel("State/UT")
plt.ylabel("Bed Gap")
plt.title("Bed Gap in Hospital Facilities")
plt.xticks(rotation=90)
plt.tight_layout()

plt.show()
```
